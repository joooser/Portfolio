# Lets make some Graphs to see the data distribution.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize=(6, 6))  # tamaño de la figura
fig.set_facecolor('lightgrey')  # establecer el fondo gris
plt.scatter(dataset.index, dataset['totalScore'], color='blue', s=dataset['reviewsCount'], alpha=0.35)  # scatterplot de circulos color azul
plt.xlabel('Indice')  # etiqueta del eje x
plt.ylabel('Rating')  # etiqueta del eje y
plt.title('Distribución de Reviews')  # título de la gráfica
plt.show()  # mostrar la gráfica 
```

As we may see in our distribution of data, all the restaurants in the area have a good rating, even acceptable, just one have a regular score. that means the competition is fearce for every customer in the city. and the biggest restaurants in the area have between 4.25 and 4,7 stars of rating, some with more than 2500 reviews. Our brand rates in 4.6stars in this city.

```python
fig = plt.figure(figsize=(6, 6))  # tamaño de la figura
fig.set_facecolor('lightgrey')  # establecer el fondo gris
plt.scatter(dataset['categories/0'], dataset['totalScore'], color='blue', s=dataset['reviewsCount'], alpha=0.4)  # scatterplot de circulos color azul
plt.ylabel('Rating')  # etiqueta del eje y
plt.title('Categorias')  # título de la gráfica

# rotar los nombres de las categorías en el eje x y ajustar el ancho de las etiquetas
plt.xticks(rotation=90)
#plt.subplots_adjust(bottom=0.25)

plt.show()  # mostrar la gráfica
```

As we may see the generic category of 'restaurant' is the most crowded, and the 5 star reviews, are all very low total count, so the most real ones oscilates between 4.25-4.75 stars being the ones with mosts reiews the pizza restaurants and the fast food restaurants
